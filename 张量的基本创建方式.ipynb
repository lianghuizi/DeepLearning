```python
import torch
import numpy as np
```

张量的基本创建方式：
  torch.tensor 根据指定数据创建张量
  torch.Tensor:根据形状创建张量，也可以用来创建指定数据的张量
  torch.IntTensor、torch.FloatTensor、torch.DoubleTensor创建指定类型的张量

细节：
  Tensor方式较之于tensor方式，可以基于 形状来直接创建张量

```python
def dm01():
  #标量转成张量
  t1=torch.tensor(10)
  print(f't1:{t1},type:{type(t1)}')

  print('-'*30)
  #列表转成张量
  data=[[1,2,3],[3,4,5]]
  t2=torch.tensor(data)
  print(f't2:{t2},type:{type(t2)}')

  print('-'*30)
  #numpy nd数组转换成张量
  data=np.random.randint(0,10,size=(2,3))
  t3=torch.tensor(data,dtype=torch.float)
  print(f't3:{t3},type:{type(t3)}')

  print('-'*30)
  #尝试直接创建 指定维度（2行3列）的张量
  #t4=torch.tensor(2,3)       #报错
  #print(f't4:{t4},type:{type(t4)}')
```

```python
def dm02():
  #标量转成张量
  t1=torch.Tensor(10)
  print(f't1:{t1},type:{type(t1)}')

  print('-'*30)
  #列表转成张量
  data=[[1,2,3],[3,4,5]]
  t2=torch.Tensor(data)
  print(f't2:{t2},type:{type(t2)}')

  print('-'*30)
  #numpy nd数组转换成张量
  data=np.random.randint(0,10,size=(2,3))
  t3=torch.Tensor(data) #前面已经制定了，这里就不让再指定了
  print(f't3:{t3},type:{type(t3)}')

  print('-'*30)
  #尝试直接创建 指定维度（2行3列）的张量
  t4=torch.Tensor(2,3)       #报错
  print(f't4:{t4},type:{type(t4)}')
```

```python
def dm03():
  #标量转成张量
  t1=torch.IntTensor(10)
  print(f't1:{t1},type:{type(t1)}')

  print('-'*30)
  #列表转成张量
  data=[[1,2,3],[3,4,5]]
  t2=torch.IntTensor(data)
  print(f't2:{t2},type:{type(t2)}')

  print('-'*30)
  #numpy nd数组转换成张量
  data=np.random.randint(0,10,size=(2,3))
  t3=torch.IntTensor(data) #前面已经制定了，这里就不让再指定了
  print(f't3:{t3},type:{type(t3)}')

  print('-'*30)
  #如果类型不匹配，会尝试自动转换类型
  data=np.random.randint(0,10,size=(2,3)) #默认float32（一个字节8个比特位，32位是4个字节，位越大范围越大）
  t4=torch.FloatTensor(data)
  print(f't4:{t4},type:{type(t4)}')
```

```python
if __name__=='__main__':
  dm01  #掌握
  # dm02()
  dm03()
```
